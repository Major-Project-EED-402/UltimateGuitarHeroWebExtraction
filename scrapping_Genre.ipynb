```python
def print_beautified_html(htmlElement):
    print(htmlElement.prettify())
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup

driver = webdriver.Chrome()
driver.get("https://www.ultimate-guitar.com/explore?type[]=Chords")
show_all_btn = driver.find_elements(By.XPATH, "//*[contains(text(), 'Show All')]")
print(show_all_btn[2])

genre_show_all = show_all_btn[2]
driver.execute_script("arguments[0].click();", genre_show_all)
```

```python
pip install lxml
```

```python
page_source = driver.page_source

soup = BeautifulSoup(page_source, "lxml")
print_beautified_html(soup)
```

```python
results = soup.findAll("div")
last_div = results[len(results)-1]
print(last_div.prettify())
```

```python
all_filter_items = last_div.find_all('a')
```

```python
print(all_filter_items)
for items in all_filter_items:
    print(items.prettify())
```

```python
# Extract genre information and store it in a list of dictionaries
genre_info_list = []
for item in all_filter_items:
    href = item.get('href')
    code = href.split('=')[-1]
    name = item.find('span', class_='wIJI1').text
    filter_pattern = f'&&genres[]={code}'
    genre_dict = {
        f'code': code, "name": name, "pattern": filter_pattern
    }
    genre_info_list.append(genre_dict)

# Print the list of dictionaries
print(genre_info_list)
```
